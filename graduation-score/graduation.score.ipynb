```python
import requests

import pandas as pd
```

```python
f = open("diemthi.txt", 'a')
```

```python
for i in range(2000001, 2085088):
    scrapping_url = "https://dantri.com.vn/thpt/1/0/99/0" + \
    str(i) + "/2022/0.2/search-gradle.htm"
    payload = {}
    headers = {}

    response = requests.request("GET", scrapping_url, headers = headers, data = payload)
    info = response.json()['student']
    diem = "{}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}".format(
        info['sbd'], info['toan'], info['van'], info['ngoaiNgu'], info['vatLy'], info['hoaHoc'], info['sinhHoc'], info['diemTBTuNhien'], info['lichSu'], info['diaLy'], info['gdcd'], info['diemTBXaHoi'])
    diemThi = str(diem) + '\n'
    f.write(diemThi)
    print(diemThi)
f.close
```
